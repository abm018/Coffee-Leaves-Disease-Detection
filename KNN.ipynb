```python
import os
import cv2
import numpy as np
```

```python
DIRECTORY = "/mydrive/PlantVillage"
CATEGORIES = ["Cerscospora", "Leaf_rust", "Phoma"]
```

```python
for category in CATEGORIES:
    path = os.path.join(DIRECTORY, category)
    for img in os.listdir(path):
      img_path = os.path.join(path, img)
      print(img_path)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn import preprocessing
import joblib
from pydrive.auth import GoogleAuth
from pydrive.drive import GoogleDrive

# Authenticate and create PyDrive client
gauth = GoogleAuth()
# gauth.LocalWebserverAuth() # Creates local webserver and automatically handles authentication.
drive = GoogleDrive(gauth)

# Function to download images from Google Drive folder
def load_images(folder):
    images = []
    labels = []
    for filename in os.listdir(folder):
        img_path = os.path.join(folder, filename)
        img = cv2.imread(img_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  # Convert to grayscale
        img = cv2.resize(img, (100, 100))  # Resize for consistency
        images.append(img.flatten())  # Flatten the image matrix
        labels.append(folder)
    return images, labels

# Google Drive folder IDs for the three diseases
cercospora_folder_id = '/mydrive/PlantVillage/Cerscospora'
leaf_rust_folder_id = '/mydrive/PlantVillage/Leaf_rust'
phoma_folder_id = '/mydrive/PlantVillage/Phoma'

# Load diseased datasets from Google Drive
cercospora_images, cercospora_labels = load_images(cercospora_folder_id)
leaf_rust_images, leaf_rust_labels = load_images(leaf_rust_folder_id)
phoma_images, phoma_labels = load_images(phoma_folder_id)
```

```python
# Concatenate datasets
all_images = np.concatenate([cercospora_images, leaf_rust_images, phoma_images])
all_labels = np.concatenate([cercospora_labels, leaf_rust_labels, phoma_labels])
```

```python
# Flatten and preprocess images
all_images = [img.flatten() for img in all_images]
all_images = preprocessing.scale(all_images)
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(all_images, all_labels, test_size=0.2, random_state=42)
```

```python
# Create a KNN classifier
knn_model = KNeighborsClassifier(n_neighbors=3)
```

```python
# Train the classifier
knn_model.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred = knn_model.predict(X_test)
```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
rounded_accuracy = round(accuracy, 6)

print(f"Accuracy: {rounded_accuracy*100}%")
```

```python
# Save the model
joblib.dump(knn_model, 'coffee_disease_model.pkl')
```
